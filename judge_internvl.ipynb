```python
import torch
from transformers import AutoTokenizer, AutoModel, BitsAndBytesConfig

model_id = "OpenGVLab/InternVL2-8B"

# Use float16 for Tesla T4 compatibility
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type='nf4'
)

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True, use_fast=False)

model = AutoModel.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    torch_dtype=torch.float16,
    trust_remote_code=True,
    device_map='auto' # Let accelerate handle optimal placement
).eval()

print("InternVL2 8B loaded successfully on T4")
```

```python
import pandas as pd
import os
from datetime import datetime
from PIL import Image
import torch
import torchvision.transforms as T
from torchvision.transforms.functional import InterpolationMode

# Image preprocessing
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD  = (0.229, 0.224, 0.225)

def build_transform(input_size=448):
    return T.Compose([
        T.Lambda(lambda img: img.convert("RGB")),
        T.Resize((input_size, input_size),
                  interpolation=InterpolationMode.BICUBIC),
        T.ToTensor(),
        T.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD)
    ])

def load_image(image_path, input_size=448):
    image = Image.open(image_path).convert("RGB")
    transform = build_transform(input_size)
    pixel_values = transform(image).unsqueeze(0).to(torch.float16).cuda()
    return pixel_values

# Judge prompt
JUDGE_PROMPT = """You are a forensic image analyst. Look at this image carefully.

A previous model analyzed this image and said:
"{gemma_reasoning}"

Based on your own visual inspection and the above analysis, is this image genuine or has it been modified?
Give your own observations in 1-2 sentences, then end with your verdict: NOT_MODIFIED or MODIFIED"""

def judge_image(image_path, gemma_reasoning):
    pixel_values = load_image(image_path)

    prompt = JUDGE_PROMPT.format(
        gemma_reasoning=gemma_reasoning[:500]
    )

    generation_config = dict(
        max_new_tokens=150,
        do_sample=False
    )

    with torch.no_grad():
        response = model.chat(
            tokenizer,
            pixel_values,
            prompt,  # no <image> tag
            generation_config
        )

    upper = response.upper()
    if "NOT_MODIFIED" in upper:
        verdict = "NOT_MODIFIED"
    elif "MODIFIED" in upper:
        verdict = "MODIFIED"
    else:
        verdict = "UNCLEAR"

    return {
        "verdict": verdict,
        "reasoning": response
    }

# Paths
GEMMA_MODEL_NAME = "gemma3-4b"
GEMMA_CSV = f"/content/drive/MyDrive/CSE_299_PROJECT/tracking/results_{GEMMA_MODEL_NAME}_fmain.csv"

REAL_DIR = "/content/drive/MyDrive/CSE_299_PROJECT/balanced/real"
FAKE_DIR = "/content/drive/MyDrive/CSE_299_PROJECT/balanced/fake"

JUDGE_MODEL_NAME = "internvl2-8b-judge"
OUTPUT_CSV = f"/content/drive/MyDrive/CSE_299_PROJECT/tracking/results_{JUDGE_MODEL_NAME}.csv"

os.makedirs(os.path.dirname(OUTPUT_CSV), exist_ok=True)

# Load Gemma CSV
gemma_df = pd.read_csv(GEMMA_CSV)
print(f"Loaded {len(gemma_df)} rows from Gemma CSV")
print(gemma_df["true_label"].value_counts())

# Run judge
judge_results = []

for i, row in gemma_df.iterrows():
    filename        = row["image_id"]
    true_label      = row["true_label"]
    gemma_verdict   = row["model_verdict"]
    gemma_reasoning = str(row["reasoning"])

    if true_label == "NOT_MODIFIED":
        image_path = os.path.join(REAL_DIR, filename)
    else:
        image_path = os.path.join(FAKE_DIR, filename)

    if not os.path.exists(image_path):
        print(f"  Image not found: {filename}, skipping")
        continue

    try:
        result = judge_image(image_path, gemma_reasoning)
        correct       = result["verdict"] == true_label
        gemma_correct = gemma_verdict == true_label

        judge_results.append({
            "image_id":        filename,
            "true_label":      true_label,
            "gemma_verdict":   gemma_verdict,
            "gemma_reasoning": gemma_reasoning,
            "judge_verdict":   result["verdict"],
            "judge_reasoning": result["reasoning"],
            "correct":         correct,
            "gemma_correct":   gemma_correct,
            "model_name":      JUDGE_MODEL_NAME,
            "timestamp":       datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })

        if (i + 1) % 10 == 0:
            print(f"  [{i+1}/{len(gemma_df)}] "
                  f"Gemma: {gemma_verdict} → "
                  f"Judge: {result['verdict']} "
                  f"({'✓' if correct else '✗'})")

        if (i + 1) % 25 == 0:
            pd.DataFrame(judge_results).to_csv(OUTPUT_CSV, index=False)
            print(f"  ✓ Checkpoint saved at {i+1}")

    except Exception as e:
        print(f"  Error on {filename}: {e}")

pd.DataFrame(judge_results).to_csv(OUTPUT_CSV, index=False)
print(f"\nDone. Saved to {OUTPUT_CSV}")

# Summary
df = pd.read_csv(OUTPUT_CSV)
total      = len(df)
judge_acc  = df["correct"].mean() * 100
gemma_acc  = df["gemma_correct"].mean() * 100
judge_real = df[df["true_label"]=="NOT_MODIFIED"]["correct"].mean() * 100
judge_fake = df[df["true_label"]=="MODIFIED"]["correct"].mean() * 100
gemma_real = df[df["true_label"]=="NOT_MODIFIED"]["gemma_correct"].mean() * 100
gemma_fake = df[df["true_label"]=="MODIFIED"]["gemma_correct"].mean() * 100

print(f"\n=== Gemma vs Judge Comparison ===")
print(f"Total             : {total}")
print(f"                    Gemma     Judge")
print(f"Overall accuracy  : {gemma_acc:.1f}%     {judge_acc:.1f}%")
print(f"NOT_MODIFIED acc  : {gemma_real:.1f}%     {judge_real:.1f}%")
print(f"MODIFIED acc      : {gemma_fake:.1f}%     {judge_fake:.1f}%")
print(f"UNCLEAR           : {(df['judge_verdict']=='UNCLEAR').sum()}")
print(f"\n=== Judge Confusion Matrix ===")
print(pd.crosstab(df["true_label"], df["judge_verdict"], margins=True))
print(f"\n=== Correction Analysis ===")
corrected  = df[(df["gemma_correct"]==False) & (df["correct"]==True)]
made_worse = df[(df["gemma_correct"]==True)  & (df["correct"]==False)]
print(f"Gemma wrong → Judge fixed : {len(corrected)}")
print(f"Gemma right → Judge broke : {len(made_worse)}")
```
